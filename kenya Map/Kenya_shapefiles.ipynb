```python
import geopandas as gpd 
import matplotlib.pyplot as plt 
```

```python
shapefiles_path = "ke_shp/ke.shp"
```

```python
kenya = gpd.read_file(shapefiles_path)
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
kenya.plot(ax=ax, color='whitesmoke', edgecolor='black', linewidth=0.5)
plt.title("Map of Kenya", fontsize=15)
plt.show()
```
