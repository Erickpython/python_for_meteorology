```python
import geopandas as gpd 
import matplotlib.pyplot as plt 
```

```python
kenya_map = gpd.read_file("ke.json")
```

```python
fig, ax = plt.subplots(figsize=(15,15))
kenya_map.plot(ax=ax, color="lightblue", edgecolor="black")
plt.title("Map of Kenya", fontsize=15)
plt.savefig("kenya_map_output.png", dpi=300, bbox_inches='tight')
plt.show()
```
