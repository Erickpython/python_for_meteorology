```python
import cdsapi
```

```python
c = cdsapi.Client()
kenya_bounds = [10, 30, -10, 45] # N, W, S, E
target_file_pl = 'era5_kenya_pl.grib'
target_file_sfc = 'era5_kenya_sfc.grib'


# 1. DOWNLOAD PRESSURE LEVELS (3D DATA)
c.retrieve('reanalysis-era5-pressure-levels', {
    'product_type': 'reanalysis',
    'format': 'grib',
    'variable': [
        'geopotential', 'relative_humidity', 'temperature',
        'u_component_of_wind', 'v_component_of_wind',
    ],
    'pressure_level': [
        '1', '2', '3', '5', '7', '10', '20', '30', '50', '70',
        '100', '125', '150', '175', '200', '225', '250', '300',
        '350', '400', '450', '500', '550', '600', '650', '700',
        '750', '775', '800', '825', '850', '875', '900', '925',
        '950', '975', '1000',
    ],
    'year': '2025',
    'month': '12',
    'day': ['01', '02'],
    'time': ['00:00', '06:00', '12:00', '18:00'],
    'area': kenya_bounds,
}, target_file_pl)

print(f"Downloading {target_file_pl} completed")
```

```python
# 2. DOWNLOAD SINGLE LEVELS (SURFACE DATA) - UPDATED
c.retrieve('reanalysis-era5-single-levels', {
    'product_type': 'reanalysis',
    'data_format': 'grib',   # Use data_format instead of format
    'variable': [
        '2m_temperature', '2m_dewpoint_temperature', '10m_u_component_of_wind',
        '10m_v_component_of_wind', 'mean_sea_level_pressure', 'surface_pressure',
        'sea_surface_temperature', 'skin_temperature', 'snow_depth',
        'volumetric_soil_water_layer_1', 'volumetric_soil_water_layer_2',
        'volumetric_soil_water_layer_3', 'volumetric_soil_water_layer_4',
        'soil_temperature_level_1',      # Changed 'layer' to 'level'
        'soil_temperature_level_2',      # Changed 'layer' to 'level'
        'soil_temperature_level_3',      # Changed 'layer' to 'level'
        'soil_temperature_level_4',      # Changed 'layer' to 'level'
    ],
    'year': '2025',
    'month': '12',
    'day': ['01', '02'],
    'time': ['00:00', '06:00', '12:00', '18:00'],
    'area': kenya_bounds,
}, target_file_sfc)

print(f"Downloading {target_file_sfc} completed")
```
